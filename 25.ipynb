```python
from aocd.models import Puzzle
from pathlib import Path

puzzle = Puzzle(year=2024, day=int(Path(__vsc_ipynb_file__).stem))

def parse_schematic(schematic_str):
    lines = schematic_str.strip().splitlines()
    return lines

def get_lock_data(lock_schematic):
    lines = parse_schematic(lock_schematic)
    num_rows = len(lines)
    if not lines:
        return [], num_rows
    num_cols = len(lines[0])
    pins = []
    for c in range(num_cols):
        height = 0
        for r in range(num_rows):
            if lines[r][c] == '#':
                height += 1
            else:
                break
        pins.append(height)
    return pins, num_rows

def get_key_data(key_schematic):
    lines = parse_schematic(key_schematic)
    num_rows = len(lines)
    if not lines:
        return [], num_rows
    num_cols = len(lines[0])
    heights = []
    for c in range(num_cols):
        height = 0
        for r in range(num_rows - 1, -1, -1):
            if lines[r][c] == '#':
                height += 1
            else:
                break
        heights.append(height)
    return heights, num_rows

def solve_a(input_data):
    parts = input_data.split('\n\n')
    lock_schematics = [p.strip() for p in parts[0].split('#####\n') if p.strip()]
    key_schematics = [p.strip() for p in parts[1].split('#####\n') if p.strip()]

    fitting_pairs = 0
    for lock_schematic in lock_schematics:
        lock_pins, lock_rows = get_lock_data(lock_schematic)
        if not lock_pins:
            continue

        for key_schematic in key_schematics:
            key_heights, key_rows = get_key_data(key_schematic)
            if not key_heights:
                continue

            if len(lock_pins) != len(key_heights):
                continue

            fits = True
            for i in range(len(lock_pins)):
                if lock_pins[i] + key_heights[i] > lock_rows:
                    fits = False
                    break
            if fits:
                fitting_pairs += 1

    return fitting_pairs

puzzle.answer_a = solve_a(puzzle.input_data)
```

```python
from aocd.models import Puzzle
from pathlib import Path

puzzle = Puzzle(year=2024, day=int(Path(__vsc_ipynb_file__).stem))

def parse_block(block_str):
    return [line.strip() for line in block_str.strip().splitlines()]

def get_lock_pins(lock_lines):
    num_rows = len(lock_lines)
    num_cols = len(lock_lines[0])
    pins = []
    for c in range(num_cols):
        height = 0
        for r in range(num_rows):
            if lock_lines[r][c] == '#':
                height += 1
            else:
                break
        pins.append(height)
    return pins

def get_key_heights(key_lines):
    num_rows = len(key_lines)
    num_cols = len(key_lines[0])
    heights = []
    for c in range(num_cols):
        height = 0
        for r in range(num_rows - 1, -1, -1):
            if key_lines[r][c] == '#':
                height += 1
            else:
                break
        heights.append(height)
    return heights

def solve_a(input_data):
    blocks = [block.strip() for block in input_data.split('\n\n') if block.strip()]
    locks_data = []
    keys_data = []

    for block_str in blocks:
        lines = parse_block(block_str)
        if lines and len(lines) == 7 and len(lines[0]) == 5:
            if lines[0] == '#####':
                locks_data.append(get_lock_pins(lines))
            elif lines[-1] == '#####':
                keys_data.append(get_key_heights(lines))

    fitting_pairs = 0
    for lock_pins in locks_data:
        for key_heights in keys_data:
            fits = True
            for i in range(len(lock_pins)):
                if lock_pins[i] + key_heights[i] > 7:
                    fits = False
                    break
            if fits:
                fitting_pairs += 1

    return fitting_pairs

puzzle.answer_a = solve_a(puzzle.input_data)
```
